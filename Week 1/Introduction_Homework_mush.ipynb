## Task 1:- Find the box[20 points]

#### Suppose that we have three colored boxes r (red), b (blue), and g (green).
#### - Box r contains 3 apples, 4 oranges, and 3 limes, 
#### - Box b contains 1 apple, 1 orange, and 0 limes, and 
#### - Box g contains 3 apples, 3 oranges, and 4 limes. 
#### soln1: If a box is chosen at random with probabilities p(r) = 0.2, p(b) = 0.2, p(g) = 0.6, and a piece of fruit is removed from the box (with equal probability of selecting any of the items in the box), then what is the probability of selecting an apple? 
#### soln2: If we observe that the selected fruit is, in fact an orange, what is the probability that it came from the green box?

**Given:**
$p(r) = 0.2, p(b) = 0.2, p(g) = 0.6$

**soln1:**
\begin{align*}
p(\text{selecting an apple}) = p(\text{apple}) &= p(r) \times  \frac{\text{apples in red bag}}{\text{fruits in red bag}} + p(b) \times  \frac{\text{apples in blue bag}}{\text{fruits in blue bag}} + p(g) \times  \frac{\text{apples in green bag}}{\text{fruits in green bag}} \\
&= 0.2 \times  \frac{3}{10} + 0.2 \times  \frac{1}{2} + 0.6 \times  \frac{3}{10} \\
&= 0.34
\end{align*}

**soln2:**

\begin{align*}
p(\text{selecting an orange}) = p(\text{orange}) &= p(r) \times  \frac{\text{oranges in red bag}}{\text{fruits in red bag}} + p(b) \times  \frac{\text{oranges in blue bag}}{\text{fruits in blue bag}} + p(g) \times  \frac{\text{oranges in green bag}}{\text{fruits in green bag}} \\
&= 0.2 \times  \frac{3}{10} + 0.2 \times  \frac{1}{2} + 0.6 \times  \frac{3}{10} \\
&= 0.34
\end{align*}

\begin{align*}
p(\text{green}\mid\text{orange}) &=  \frac{p(g) \times p(\text{oranges} \mid \text{green})}{p(\text{orange})} \\
&=  \frac{0.2 \times 0.6}{0.36} \\
&= 0.5
\end{align*}

```python
# probabilities of choosing a *color* bag:
p_r = 0.2
p_b = 0.2
p_g = 0.6

p_r_a = 3
p_r_o = 4
p_r_l = 3

p_b_a = 1
p_b_o = 1
p_b_l = 0

p_g_a = 3
p_g_o = 3
p_g_l = 4

p_apple = (p_r * (p_r_a / (p_r_a + p_r_o + p_r_l))) \
        + (p_b * (p_b_a / (p_b_a + p_b_o + p_b_l))) \
        + (p_g * (p_g_a / (p_g_a + p_g_o + p_g_l)))

p_orange_given_green = (p_g * (p_g_o / (p_g_a + p_g_o + p_g_l)))

soln1 = p_apple
soln2 = p_orange_given_green

# soln1, soln2
```

## Task 2:- Let's get into magic[20 points]

#### - the second card is colored green on one side and blue on the other side 
#### - the third card is colored blue on both sides. 

#### Now one card is randomly selected, which has green color on one side. What is the probability that the other side is blue?

##### **Hint**: Monty Hall problem


Given that the side of the card picked is **Green** there are three equally likely events:

1. You pick a **Green** card which is **Blue** on the other side which has a probability of 1/3
2. You pick a **Green** card which is **Green** on the other side which is **twice** as likely to happen with a probablity of 2/3. Here:
         - In the first case: Pick a green card side 1 and flip to side 2 which is also green
         - In the second case: Pick a green card side 2 and flip to side 1 which is also green
         
Thus required **soln3**: probability of picking a card with blue color on the other side given that the color on the present side is green is 1/3.

```python
p_other_side_green_given_green = 1/3 + 1/ 3
p_other_side_blue_given_green = 1 - p_other_side_green_given_green

soln3 = p_other_side_blue_given_green
```

## Task 4:- Gaussian Distribution[40 points]
#### Your final output should look like below 


#### Consider a Gaussian distribution $p_x(x)$ over x with mean μ = 6 and standard deviation $\sigma$ = 1. Plot this distribution.

```python
mu = 6
sigma = 1
x = np.linspace(mu - 3 * sigma, mu + 3 * sigma, 100)
x = np.linspace(0, 10, 100)
guassian_dist = lambda x: np.exp(-np.square(x - mu) / 2 * sigma) / (np.sqrt(2 * np.pi * sigma))

# Plotting
# plot should have axis labels, grid, legend, title, atleast size 10X10
# Also give proper comments, function name, 
# variable names to your coding, if you didn't follow the instructions 
# there will be a reduction in the points.

plt.figure(figsize= (10, 10))
plt.axis([0,11,0,1.0])
plt.plot(x, guassian_dist(x))
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Guassian Distribution')
plt.legend(['p(x)'])
plt.grid()
plt.show()

# References:
# https://stackoverflow.com/questions/10138085/python-plot-normal-distribution
```

#### Draw a sample of N = 50000 points from this distribution and plot a histogram of their values, which as expected agrees with the distribution $p_x(x)$. 

```python
import numpy as np
import matplotlib.pyplot as plt


s = np.random.normal(mu, sigma, 50000)
num_bins = 30

plt.figure(figsize= (10, 10))
plt.axis([0,11,0,1.0])
plt.plot(x, guassian_dist(x))
plt.hist(s, bins=num_bins, facecolor='blue', alpha=0.5, density=True, edgecolor = 'black')

plt.xlabel('x')
plt.ylabel('p(x)')
plt.title('Histogram of the random samples drawn from p(x)')
plt.legend(['p(x)','histogram'])
plt.grid()
plt.show()

# plt.plot(guassian_dist)
```
