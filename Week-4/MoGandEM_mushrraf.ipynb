```python
#https://www.visiondummy.com/2014/04/draw-error-ellipse-representing-covariance-matrix/
#https://www.xarg.org/2018/04/how-to-plot-a-covariance-error-ellipse/    
def plotEllipsoids(cov_mat, centre, nstd, facecolor='none',**kwargs):
    # Find and sort eigenvalues and eigenvectors into descending order
    eigvals, eigvecs = np.linalg.eigh(cov_mat)
    order = eigvals.argsort()[::-1]
    eigvals, eigvecs = eigvals[order], eigvecs[:, order]
    # The anti-clockwise angle to rotate our ellipse by
    vx, vy = eigvecs[:,0][0], eigvecs[:,0][1]
    theta = np.arctan2(vy, vx)

    # Width and height of ellipse to draw #The eigenvalues represent the spread of the data in the direction of the eigenvectors.
    width, height =2 * nstd * eigvals
    return Ellipse(xy=centre, width=width, height=height,
                   angle=np.degrees(theta), facecolor = facecolor, **kwargs)

def algorithmEM(data_set):
    #function for EM algorithm
    #storing the data_set in data variable
    data = data_set
    #creating a data frame to help us in assigning label name
    data_frame = pd.DataFrame(data,columns = ['x', 'y'])
    #iniitilazing sklearn gaussian mixture with 3 components
    model = GaussianMixture(n_components = 3, init_params= 'kmeans',
                            covariance_type='full', random_state=100,)
    #fittign the model on the data
    model.fit(data)
    #finding the probability scores over all components
    probability_scores = model.predict_proba(data)
    #doing soft assigment of cluster if the probability is greater than 0.9
    list_label = []
    for index in range(probability_scores.shape[0]):
        max_cluster = np.argmax(probability_scores[index,:])
        if probability_scores[index, max_cluster] > 0.9:
            list_label.append(max_cluster)
        else:
            list_label.append('outliers')

    data_frame['labels'] = list_label

    #plotting the points only
    fig, ax  = plt.subplots(1, 1, figsize = (10,10))
    ax.scatter(data_frame.loc[:,'x'],
               data_frame.loc[:,'y'], color = 'black')
    ax.set_title("Initial distribution")
    ax.set_xlabel('x_axis')
    ax.set_ylabel('y_axis')
    plt.grid()
    plt.show()
    
    #plotting soft assignment with ellipsoid
    fig, ax  = plt.subplots(1, 1, figsize = (10,10))
    #plotting outliers
    ax.scatter(data_frame.loc[data_frame.labels == 'outliers','x'],
               data_frame.loc[data_frame.labels == 'outliers','y'],
              label = "outliers", color = 'orange')
    #plotting cluster point and their outliers
    for cluster,col in zip([0,1,2],['blue','red','purple']):
        #filtering for the cluster
        data_sub = data_frame.loc[data_frame.labels == cluster,:]
        #plotting cluster point
        ax.scatter(data_sub.values[:,0],data_sub.values[:,1],
              label = "Cluster "+ str(cluster), color = col)
        #plotting different standard deviation clusters
        for i in [1,2,3]: #1sd = 68%, 2sd = 95%, 3sd = 99% 
            elli = plotEllipsoids(np.cov(data_sub.values[:,:2].T.astype(float)),
                                  np.mean(data_sub.values[:,:2].astype(float),axis = 0),
                                  i,edgecolor=col,
            label = str(i) + 'confidence')
            #adding it to the already existing ax
            ax.add_artist(elli)
    ax.set_xlabel('x_axis')
    ax.set_ylabel('y_axis')
    plt.grid()
    plt.legend()
    plt.title("Soft assignment and Ellipsoid")
    #plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    
    
    #### plotting hard assignment
    
    #again making a data frame
    data_frame = pd.DataFrame(data,columns = ['x', 'y'])

    #initializing gaussian
    model = GaussianMixture(n_components = 3, init_params= 'kmeans',
                            covariance_type='full', random_state=100)
    #fitting data
    model.fit(data)
    #getting probability scores for each component
    probability_scores = model.predict_proba(data)
    #doing hard assignment
    list_label = []
    for index in range(probability_scores.shape[0]):
        max_cluster = np.argmax(probability_scores[index,:])
        list_label.append(max_cluster)

    data_frame['labels'] = list_label

    #plotting hard assignment points with their elliposid
    fig, ax  = plt.subplots(1, 1, figsize = (10,10))
    
    #plotting clusters and their point

    for cluster,col in zip([0,1,2],['blue','red','purple']):
        #filtering for the cluster for the current cluster
        data_sub = data_frame.loc[data_frame.labels == cluster,:]
        #plotting points
        ax.scatter(data_sub.values[:,0],data_sub.values[:,1],
              label = "Cluster "+ str(cluster), color = col)
        #plotting ellipsoid for different standard deviations
        for i in [1,2,3]: #1sd, 2sd, 3sd
            elli = plotEllipsoids(model.covariances_[cluster],
                                  model.means_[cluster],
                                  i,edgecolor=col,
            label = str(i) + 'confidence')
            ax.add_artist(elli)
    ax.set_xlabel('x_axis')
    ax.set_ylabel('y_axis')
    plt.grid()
    plt.legend()
    plt.title("Hard Assignments of points and ellipsid")
    #plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

    return

if __name__ == '__main__':
    algorithmEM(data)
```

#### Explain the concept behind your ellipsoid plotting, like how you plotted it [5 points]

```python
def plotEllipsoids(cov_mat, centre, nstd, facecolor='none',**kwargs):
    # Find and sort eigenvalues and eigenvectors into descending order
    eigvals, eigvecs = np.linalg.eigh(cov_mat)
    order = eigvals.argsort()[::-1]
    eigvals, eigvecs = eigvals[order], eigvecs[:, order]
    # The anti-clockwise angle to rotate our ellipse by
    vx, vy = eigvecs[:,0][0], eigvecs[:,0][1]
    theta = np.arctan2(vy, vx)

    # Width and height of ellipse to draw #The eigenvalues represent the spread of the data in the direction of the eigenvectors.
    width, height =2 * nstd * eigvals
    return Ellipse(xy=centre, width=width, height=height,
                   angle=np.degrees(theta), facecolor = facecolor, **kwargs)
```

- For the plotting of ellipsoids plotEllipsoids() function is used. This function takes covariance matrix, center, and standard deviation number as the main argument
- In this funtion first, eigen values and eigen vectors of covariance matrix are found because eigen vectors have direction towards the maximum variation and eigen value represents how much variantion there is
- Then eigen vector and eigen values were sorted w.r.t the eigen values in descending order to get the maximum variation change.
- Angle of orientation of ellipse is found using the eigen vector having the largest eigen value
- width and height of the ellipse is found by multiplying the standard deviation number with eigen value because it represents the deviation.
